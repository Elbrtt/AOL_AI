```python
from flask import Flask, request
import numpy as np
from tensorflow.keras.models import load_model

app = Flask(__name__)

model = load_model("models/food_spoil_model3.h5")

@app.route('/')
def home():
    return """
    <h2>Prediksi Food Spoilage</h2>
    <form action='/predict' method='post'>
        Suhu (°C): <input name='temp' type='number' step='any'><br><br>
        Jam disimpan: <input name='hours' type='number' step='any'><br><br>
        <button type='submit'>Prediksi</button>
    </form>
    """

@app.route('/predict', methods=['POST'])
def predict():
    temp = float(request.form['temp'])
    hours = float(request.form['hours'])

    data = np.array([[temp, hours]])
    pred = model.predict(data)[0][0]

    return f"Hasil prediksi: {pred}"

if __name__ == '__main__':
    app.run(debug=False)
```
