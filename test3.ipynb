```python
import pandas as pd
from tensorflow.keras.models import load_model

# 1. Load model yang sudah disimpan
model = load_model("foodwaste_model.keras")
print("Model loaded successfully.")

# 2. Contoh data baru untuk prediksi
sample_input = pd.DataFrame([{
    "qty_prepared": 300,
    "qty_sold": 260,
    "temperature": 39,
    "customers_prediction": 900,
    "is_holiday": 1
}])

# 3. Prediksi food waste
prediction = model.predict(sample_input)[0][0]

# 4. Tampilkan hasil
print("Prediksi food waste:", round(prediction, 2))
```
